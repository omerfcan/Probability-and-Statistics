### Durum 1

Atilan adil bir paranin `Tura` gelme ihtimali.

Burada "adil bir paranin atilmasi" deneyini `random.choice(['Y', 'T'])` ile temsil edecegiz.
Olay ise `random.choice(['Y', 'T']) == 'T'`

```python
import random
import matplotlib.pyplot as plt
from typing import List
import math     # ileride matematiksel işlem yapacağımız için bu math kütüphanesi lazım.
```

```python
# Deneyin tanımı

def tek_para_deneyi() -> str:
    yuz = random.choice(['Y', 'T'])
    return yuz
```

```python
# Olayin tanimi

def tura_olayi(sonuc: str) -> bool:
    return sonuc == 'T'
```

```python
# Deneylerin icrasi

tekrar_sayisi = 500_000

deney_no = [0]
olay_frek = [0]
nispi_frek = [0]

for i in range(1, tekrar_sayisi + 1):
    sonuc = tek_para_deneyi()
    olay_g = int(tura_olayi(sonuc))  # int, {0, 1}
    deney_no.append(i)
    olay_frek.append(olay_frek[-1] + olay_g)
    nispi_frek.append(olay_frek[-1] / deney_no[-1])

plt.axhline(0.5, color='red', linestyle='--')
plt.plot(deney_no[1:], nispi_frek[1:])
plt.semilogx()     # Grafiğin görünümünün daha anlamlı olması için örneklemin logaritmasını almak gibi bir fonksiyon.
plt.title("Atilan paranin tura gelmesi olayi")
plt.xlabel("Tekrar sayisi")
plt.ylabel("Nispi frekans")
```

### Durum - 2

Atilan 8 paradan en az 3'unun Tura gelme ihtimali.

Bu kez deney "8 para atilmasi", olay ise "en az 3 tura gelmesi".

```python
# Deney tanimi
def cok_para_deneyi() -> List[str]:
    yuzler = []    
    for i in range(1,9):
        yuz = random.choice(["Y", "T"])
        yuzler.append(yuz)
    return yuzler
```

```python
cok_para_deneyi()
```

```python
def en_az_3_tura_olayi(sonuc: List[str]) -> bool:
    tura_sayisi = sonuc.count("T")
    return tura_sayisi >= 3
```

```python
en_az_3_tura_olayi(cok_para_deneyi())
```

```python
#Deneylerin icra edilmesi
tekrar_sayisi = 500_000

deney_no = [0]
olay_frek = [0]
nispi_frek = [0]

for i in range(1, tekrar_sayisi + 1):
    sonuc = cok_para_deneyi()
    olay_g = int(en_az_3_tura_olayi(sonuc))  # int, {0, 1}
    deney_no.append(i)
    olay_frek.append(olay_frek[-1] + olay_g)
    nispi_frek.append(olay_frek[-1] / deney_no[-1])
```

```python
nispi_frek[-1]
```

```python
# Gorsellestirme
plt.figure()
plt.axhline(0.85546875, color='red', linestyle='--')
plt.plot(deney_no[1:], nispi_frek[1:])
plt.semilogx()
plt.title("Atilan 8 paradan en az 3'unun tura gelmesi olayi")
plt.xlabel("Tekrar sayisi")
plt.ylabel("Nispi frekans")
```
